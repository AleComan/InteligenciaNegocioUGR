```python
import h2o
from h2o.automl import H2OAutoML
import pandas as pd
```

#### Iniciar h2o

```python
h2o.init()
```

#### Cargar datos

```python
h2o.estimators.xgboost.H2OXGBoostEstimator.available()
```

```python
train = h2o.import_file("train.csv")
test = h2o.import_file("test.csv")
```

### Definir variable objetivo y resto de variables

```python
y = "price"
x = train.columns
x.remove(y)
```

### Entrenar modelo

```python
aml = H2OAutoML(max_runtime_secs=5000, seed=42)
aml.train(x=x, y=y, training_frame=train)
```

```python
# Ordenar la tabla por RMSE
leaderboard_sorted = aml.leaderboard.as_data_frame().sort_values(by='rmse')
print(leaderboard_sorted)

leaderboard_sorted.to_csv("leaderboard.csv", index=False)
```

```python
model_id = 'StackedEnsemble_Best1000_1_AutoML_1_20241231_23549'
model = h2o.get_model(model_id)

print(model)
```

```python
predictions = aml.leader.predict(test)
```

```python
predictions_df = predictions.as_data_frame()
test_ids = test["id"].as_data_frame()

submission = pd.DataFrame({
    'id': test_ids['id'],
    'price': predictions_df['predict']
})

# Guardar el archivo de salida
submission_path = 'submission_h2o_3.csv'  # Ruta del archivo de salida
submission.to_csv(submission_path, index=False)
```
